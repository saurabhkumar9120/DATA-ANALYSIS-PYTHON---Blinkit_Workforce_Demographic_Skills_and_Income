# *Income Group Distribution*.

```python
# Convert Income to numeric
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')

# Define income bins and labels
bins = [0, 20000, 50000, 100000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']

# Create income group categories
df['Income_Group'] = pd.cut(df['Income'], bins=bins, labels=labels)

# Count frequency of each group
Income_Group_Count = df['Income_Group'].value_counts().sort_index()

# Define a different color for each income group
group_colors = ['#4daf4a', '#377eb8', '#ff7f00', '#e41a1c']  # green, blue, orange, red

# Plot histogram (bar chart)
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(Income_Group_Count.index, Income_Group_Count.values,
              color=group_colors, edgecolor='black')

# Add value labels on top of bars
for i, value in enumerate(Income_Group_Count.values):
    ax.text(i, value + 5, str(value), ha='center', fontsize=10)

# Format plot
ax.set_title('Income Group Distribution (Histogram)', fontsize=14)
ax.set_xlabel('Income Group', fontsize=12)
ax.set_ylabel('Number of People', fontsize=12)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# *Incom Race Bar Sumulation*.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Step 1: Prepare data - Top 10 Income
income_counts = df['Income'].value_counts().head(20).sort_values(ascending=True)
labels = income_counts.index.astype(str)
values = income_counts.values

# Step 2: Setup plot
fig, ax = plt.subplots(figsize=(8, 5))
bar_container = ax.barh(labels, [0]*len(values), color='skyblue')
ax.set_xlim(0, max(values) * 1.2)
ax.set_title("Top 10 Income (Race Bar Simulation)")
ax.set_xlabel("Race Bar")

# Step 3: Animation function
def update(frame):
    for bar, value in zip(bar_container, values):
        bar.set_width(value * (frame / 20))
    return bar_container

# Step 4: Animate over 20 frames
ani = animation.FuncAnimation(fig, update, frames=21, interval=100, blit=True)

# Step 5: Display in notebook
plt.close()  # prevent duplicate static plot
HTML(ani.to_jshtml())
```

# *Histogram: Shows how individuals’ ages (from the Age column) are diributed across the dataset*.

# *Line Chart: Displays the trend of average Income over the Years (from the Year and Income columns)*.
# *Box Plot: Highlights how Income varies across different States (from the State and Income columns), including medians and outliers*.

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # 2 rows, 2 columns

# 1. Histogram: Age Distribution
axs[0, 0].hist(df['Age'].dropna(), bins=10, color='orange', edgecolor='black')
axs[0, 0].set_title('Age Distribution')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Count')

# 2. Bar Chart: Average Income by Gender

df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
income_by_gender = df.groupby('Gender')['Income'].mean()

bars = axs[0, 1].bar(income_by_gender.index, income_by_gender.values, color='teal')
axs[0, 1].set_title('Average Income by Gender')
axs[0, 1].set_xlabel('Gender')
axs[0, 1].set_ylabel('Average Income')

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    axs[0, 1].text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:.0f}',  # format as integer
        ha='center',
        va='bottom',
        fontsize=10,
        color='black'
    )

# 3. Line Chart: Average Income by Year
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
income_by_year = df.groupby('Year')['Income'].mean().sort_index()
axs[1, 0].plot(income_by_year.index, income_by_year.values, color='crimson', linewidth=2)
axs[1, 0].set_title('Average Income Trend by Year')
axs[1, 0].set_xlabel('Year')
axs[1, 0].set_ylabel('Average Income')

# 4. Box Plot: Income Distribution by State
df.boxplot(column='Income', by='State', ax=axs[1, 1])
axs[1, 1].set_title('Income by State')
axs[1, 1].set_xlabel('State')
axs[1, 1].set_ylabel('Income')
axs[1, 1].tick_params(axis='x', rotation=45)
fig.suptitle('')  # Remove default boxplot title

plt.tight_layout()
plt.show()
```
