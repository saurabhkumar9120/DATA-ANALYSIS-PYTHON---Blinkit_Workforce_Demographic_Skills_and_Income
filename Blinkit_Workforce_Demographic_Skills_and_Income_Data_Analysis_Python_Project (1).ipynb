### **Creating Charts for Data Visualization**

```python
# creates a pivot table summing Income per Year and plots it as a bar chart using pandas’ matplotlib
pd.pivot_table(df,index = 'Year',values='Income',aggfunc='sum').plot(kind='bar')
```

```python
# groups data by Education, calculates average Income per group via pivot_table
pd.pivot_table(df, index='Education', values='Income', aggfunc='mean').plot(kind='bar')
```

```python
# creates a pivot table that counts non-null Skill_1 entries per State
pd.pivot_table(df, index='State', values='Skill_1', aggfunc='count').plot(kind='bar')
```

```python
# groups rows by State, counts non-null Skill_2 entries per state using pivot_table and plots 
pd.pivot_table(df, index='State', values='Skill_2', aggfunc='count').plot(kind='bar')
```

```python
# groups data by Industry, calculates the average Experience_Years per industry using pivot_table and plots
pd.pivot_table(df, index='Industry', values='Experience_Years', aggfunc='mean').plot(kind='bar')
```

```python
# groups rows by unique Age, counts the number of entries in the Gender column per age using pivot_table and plot 
pd.pivot_table(df, index='Age', values='Gender', aggfunc='count').plot(kind='bar')
```

```python
# counts non-null 'Education' entries for each Government_Scheme category using pivot_table + ‘count’ then plots a bar chart for scheme-wise education 
pd.pivot_table(df,index ='Government_Scheme', values='Education', aggfunc='count').plot(kind='bar')
```

```python
 # groups data by State, computes each state's average Income, sorts states descending by that average, and visualizes rankings
df.groupby('State')['Income'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')  # convert 'Income' entries to numeric
df = df.dropna(subset=['Income', 'Year']) # drop rows missing either 'Income' or 'Year'
df['Year'] = df['Year'].astype(int)  # explicitly convert 'Year' column to integer
pivot = pd.pivot_table(df, index='Year', values='Income', aggfunc='sum') # aggregate total Income per Year
pivot.plot(kind='bar', title="Total Income by Year", figsize=(10, 6)) # plot the yearly total income
plt.xlabel("Year") # label the x-axis as 'Year'
plt.ylabel("Total Income")  # label the y-axis as 'Total Income'
plt.grid(True) # enable gridlines for easier visual reading of bar heights
plt.tight_layout() # adjust layout to prevent clipping of labels and title
plt.show() # display the final bar chart showing total income by year
```

```python
Year_Count = df['Year'].value_counts().sort_index()  # count records per year and sort years in ascending order
Year_Count.plot(kind='bar', figsize=(10,5), title='Data Records by Year')  # plot the counts as a vertical bar chart
plt.xlabel('Year')  # label x-axis
plt.ylabel('Count')  # label y-axis
plt.show()  # display the plot
```

```python
# Create a binary column: 1 if Training Received == 'Yes', else 0
df['Training_Binary'] = df['Training_Received'].map({'Yes': 1, 'No': 0})

# Calculate percentage of people trained each year
training_rate = df.groupby('Year')['Training_Binary'].mean() * 100

# Plot
training_rate.plot(kind='bar', figsize=(10, 5), title='Training Participation Rate by Year')
plt.ylabel('% Trained')
plt.xlabel('Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.groupby('State')['Income'].mean().sort_values(ascending=False).plot(  # group rows by State, compute mean Income per group
    kind='bar',                                                              # specify a vertical bar chart
    title='Average Income per State',                                       # set the plot title
    figsize=(10, 6)                                                         # define figure size in inches (width, height)
)

plt.ylabel('Average Income')  # label the y-axis for clarity 
plt.xlabel('State')           # label the x-axis with the grouping variable
plt.grid(True)                # enable gridlines to improve readability 
plt.tight_layout()            # auto-adjust layout so labels/title fit without overlap
plt.show()                    # render the completed plot on screen

print(df.columns)            # display all DataFrame column names as an Index object
```

```python
Income_Count = df['Income'].value_counts().head(10)  # count unique Income values, take top 10 most frequent
Income_Count.plot(kind='bar', figsize=(8, 5), title='Top 10 Income Frequencies')  # plot frequencies as a bar chart
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.tight_layout()
plt.grid(True)
plt.show()


```

```python
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')  # convert Income to numeric

bins = [0, 20000, 50000, 100000, float('inf')]  # define numeric boundaries for income categories
labels = ['Low', 'Medium', 'High', 'Very High']  # assign descriptive labels for each income interval
df['Income_Group'] = pd.cut(df['Income'], bins=bins, labels=labels)  # bin continuous Income into labeled categories via pd.cut 

Income_Group_Count = df['Income_Group'].value_counts()  # count occurrences of each income group to measure distribution

plt.pie(Income_Group_Count, labels=Income_Group_Count.index, autopct='%1.1f%%', startangle=90)  # plot pie chart, showing percentages, rotated to start at 12 o'clock 
plt.title('Income Group Distribution')  # set the title of the pie chart
plt.axis('equal')  # ensure aspect ratio is equal so pie renders as a circle
plt.tight_layout()  # adjust layout to prevent label overlap
plt.show()  # display the completed pie chart
```

```python
Income_Count = df['Income'].value_counts().head(10)  # count unique Income values, then select top 10 most frequent entries using head(10) 
 # draw a pie chart of top 10 Income counts, label slices, show percentages with one decimal, and rotate start by 30° counterclockwise
plt.pie(Income_Count, labels=Income_Count.index, autopct='%1.1f%%', startangle=30)  

plt.title('Top 10 Income Frequencies')  # set the pie chart title for context

plt.axis('equal')  # enforce equal aspect ratio so the pie is rendered as a perfect circle

plt.tight_layout()  # auto-adjust figure layout to prevent label/title clipping

plt.show()  # display the final plot on screen
```

```python
Year_Count = df['Year'].value_counts().head(2025) # count occurrences of each Year (largest first), then take up to 2025 unique years
# draw pie chart of year frequencies, label slices, display percentages, start at 90°
plt.pie(Year_Count, labels=Year_Count.index, autopct='%1.1f%%', startangle=90)  
plt.title('2000 - 2025 Year Frequencies') # set the pie chart title for context
plt.axis('equal')  # equal aspect ratio to render pie as a circle
plt.tight_layout() # auto-adjust layout to prevent label/title clipping 
plt.show()     # display the final pie chart
```
